### Read data from income.csv and scale the relevant features using standardization

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from google.colab import drive
drive.mount('/content/drive')
data_set = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/2024-05-22T16-53-02.667Z-Scaling_Assignment/Scaling_Assignment/income.csv")
#data_set
data_set.head()
```

```python
data_set.info()
```

```python
x = data_set[['Age', 'Income']]
x
```

```python
sns.distplot(x['Age'])
```

```python
sns.distplot(x['Income'])
```

```python
x_temp = x.copy()
x_temp['Age'] = (x_temp['Age'] - x_temp['Age'].mean()) / x_temp['Age'].std()
x_temp
```

```python
sns.distplot(x_temp['Age'])
```

```python
sns.distplot(x_temp['Income'])
```

```python
from sklearn.preprocessing import StandardScaler as SC

std = SC()

x_after_Standardisation = std.fit_transform(x)

x_after_Standardisation
```

```python
x_age=sns.distplot(x_after_Standardisation[:,0])
plt.xlabel('Age')
x_age
```

```python
x_age=sns.distplot(x_after_Standardisation[:,1])
plt.xlabel('Income')
x_age
```
