Match the output below:

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y, color="green", label="y=x*2")
ax1.set_xlabel("x_label")
ax1.set_ylabel("y_labe")
ax1.set_title("Plot_titlet")
ax1.legend(loc=4)
plt.show()
```

**2. Instructions**<br>
* ** Create a figure object**
* ** Put two axes on the figure at [0,0,1,1] and [0.1,0.6,0.2,0.2]**
* ** Plot data (x,z) on main axes**
* ** Plot data (x,z) on the small axes and set its x limit b/w 1, 2, y limit 0, 5**
* ** Replicate the figure to axes, titles, legends etc**

```python
#Code here
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.6,0.2,0.2])

ax1.plot(x,z, color="red", label="y_main=x_main**2")
ax1.set_xlabel("x_main")
ax1.set_ylabel("y_main")
ax1.set_title("Title_main")
ax1.legend(loc=4)


ax2.plot(x,z, color="blue")
ax2.set_xlim(1,2)
ax2.set_ylim(0,7,2.5)
ax2.set_xlabel("x_inset")
ax2.set_ylabel("y_inset")
ax2.set_title("inset")

plt.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.1,0.6,0.2,0.2])

ax1.plot(x,z, color="red", label="y_main=x_main**2")
ax1.set_xlabel("x_main")
ax1.set_ylabel("y_main")
ax1.set_title("Title_main")
ax1.legend(loc=4)


ax2.plot(x,z, color="blue")
ax2.set_xlim(1,2)
ax2.set_ylim(0,7,2.5)
ax2.set_xlabel("x_inset")
ax2.set_ylabel("y_inset")
ax2.set_title("inset")

plt.show()
```
