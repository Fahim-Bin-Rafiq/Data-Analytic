**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.set_style('white')
sns.boxplot(x="day", y="tip", data=tips,
           hue = 'sex', palette='viridis', fliersize=5, flierprops=dict(marker='d', markerfacecolor='black'))
```

**Recreate the plot below. Use tips dataset. Hint: pay attention to x and y labels.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

**Recreate the distplot given below. Use titanic dataset. Hint: pay attention to x and y labels.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.set_style('white')
sns.distplot(titanic['fare'],kde=False,bins=30, color='green')
```

**Recreate the boxplot below using titanic dataset. Hint: pay attention to x and y labels.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.set_style('white')
sns.boxplot(x="class", y="age", data=titanic,
           hue = 'alive', palette='coolwarm',fliersize=5, flierprops=dict(marker='d', markerfacecolor='black'))
```

**Recreate the plot below using titanic dataset. Hint: pay attention to x and y labels.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.jointplot(x='fare',y='age',data=titanic)
```

**Recreate the plot below. Use titanic dataset. **

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.boxplot(x="class", y="age", data=titanic, palette='rainbow',fliersize=5, flierprops=dict(marker='d', markerfacecolor='black'))
```

**Can you plot how many male and female were on titanic? Separate them based on their class.**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
sns.set_style('white')
datalabel = sns.countplot(x="sex", hue="class",data=titanic, palette= ['blue', 'green', 'red'])
```

**Recreate the plot below. Use titanic dataset. Hint: Matrix plots**

```python
# Code here and leave the next cell empty, so that you don't lose the plot!
titanic2 = titanic[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare','adult_male', 'alone' ]]
tc = titanic2.corr()
plot = sns.heatmap(tc, cmap='coolwarm', annot = True )
plot.set_title("titanic.corr()")
```
